```python
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris[['species']]
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_tain, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)


```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='auto')

clf.fit(X_train, y_tain)
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.classes_
```

```python
np.round(clf.predict_proba(X_train),3)
```

```python
clf.score(X_train, y_tain)
```

```python
clf.score(X_test, y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn import model_selection 
from sklearn.model_selection import cross_val_score

kfold = model_selection.KFold(n_splits=10)
clf_model = LogisticRegression()
results = cross_val_score(clf_model, X_train, y_tain, cv=kfold, scoring='accuracy')
```
