## Line Chart

```python

df = pd.DataFrame()
df['Date'] = pd.date_range('01-01-2025', '03-31-2025', freq='D')
df['India Price'] = np.abs(np.random.randn(90))
df['Aus Price'] = np.abs(np.random.randn(90))
```

```python
df2 = df.copy()
```

```python
df2
```

```python

plt.figure(figsize=(15,5))
plt.plot(df2['Date'], df['India Price'])
plt.title('Daily Prices($)')
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.show()

```

```python
df2['India Price'] = df2['India Price'] * 1.2
df2['Aus Price'] = df2['India Price'] * 0.4
```

```python
plt.figure(figsize=(15,5))
plt.plot(df2['Date'], df2['India Price'])
plt.plot(df2['Date'], df2['Aus Price'])
plt.title('Daily Prices($)')
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.show()
```

```python
df = pd.DataFrame()
df['Date'] = pd.date_range(start='01-01-2025 00:00:00', end='01-07-2025 00:00:00', freq='H')
len_size = df.shape[0]
df['Temperature'] = np.abs(np.random.randn(len_size))


df2 = df.copy()

df2
```

```python
plt.figure(figsize=(15,5))
plt.plot(df2['Date'], df['Temperature'])
plt.title('Daily Prices($)')
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.show()
```

```python

points = 50
time = np.linspace(0, 10, points)
temp = np.random.uniform(low=22, high=24, size=points)

plt.figure(figsize=(15,5))
plt.plot(time, temp)
plt.title('Daily Prices($)')
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()
```

```python
points = 50
time = np.linspace(0, 10, points)
temp = np.sin(time)
temp_1  = np.sin(time) + 1.2
temp_2  = np.sin(time) + 0.6
plt.figure(figsize=(15,5))
plt.plot(time, temp)
plt.plot(time, temp_1, color='r', linestyle="--")
plt.plot(time, temp_2)
plt.title('Daily Prices($)')
plt.grid(True)
plt.legend(['Sin(x)', 'Sin(x)+1.5','Sin(x)+0.75'])
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x=time, y=temp)
plt.grid(True)
plt.show()
```
